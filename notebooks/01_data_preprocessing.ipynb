# 01 - Data Preprocessing

Loads `patients` and `admissions` from the MIMIC-IV release and builds the
modeling cohort. Run this notebook on a Databricks cluster with the data
mounted at the path configured in `config/config.yaml`.

```python
from src.utils import get_spark, load_config
from src import data_preprocessing as dp

cfg = load_config('../config/config.yaml')
spark = get_spark()
base = cfg['data']['mimic_path']
```

```python
patients = dp.load_table(spark, base, 'patients.csv')
admissions = dp.load_table(spark, base, 'admissions.csv')
cohort = dp.build_cohort(patients, admissions)
cohort.cache()
print(f'Cohort size: {cohort.count():,}')
cohort.groupBy('gender').count().show()
```

```python
cohort.write.mode('overwrite').parquet(cfg['data']['output_path'] + 'cohort/')
```
