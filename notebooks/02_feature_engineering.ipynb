# 02 - Feature Engineering

Builds the Spark ML feature pipeline (age bucketing + length-of-stay) and
writes the featurized dataset for downstream modeling.

```python
from src.utils import get_spark, load_config
from src import feature_engineering as fe

cfg = load_config('../config/config.yaml')
spark = get_spark()
cohort = spark.read.parquet(cfg['data']['output_path'] + 'cohort/')
```

```python
pipeline = fe.build_feature_pipeline()
featurized = fe.transform(cohort, pipeline)
featurized.select('subject_id', 'gender', 'label', 'features').show(5, truncate=False)
```

```python
featurized.write.mode('overwrite').parquet(cfg['data']['output_path'] + 'featurized/')
```
